## Multiple Linear Regression

Multiple Linear Regression is used when there are two or more independent variable (Input).<br>
One dependent variable (output) to predict.


## **Example in this practical:**

We want to predict Marks

Based on:

*Study Hours*

*Number of Assignments completed*

So here:

**Inputs (X):** Study_Hours, Assignments

**Output (Y):** Marks

Y=b0+b1X1+b2X2

This means:

Marks = base marks + (effect of study hours) + (effect of assignments)

Where:

b0 (Intercept): Marks when study hours and assignments are zero

b1: How much marks increase when study hours increase by 1

b2: How much marks increase when assignments increase by 1

```python
# Importing the Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Creating the Dataset
data={
    "Study Hours":[4,6,7,5,8],
    "Assignment": [3,3,4,2,5],
    "Marks":[72,80,92,74,98]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Separate input and output
X=df[["Study Hours","Assignment"]]
Y=df["Marks"]
```

```python
# Creating the model
model=LinearRegression()
model.fit(X,Y)
```

```python
# Print Intercept and Coefficient
print('Intercept(b0):,model.intercept_')
print("coefficient(b1,b2):model.coef_")
```

```python
# Predict Marks for the new students
input_df=pd.DataFrame([[7,4]],columns=["Study Hours","Assignment"])
prediction=model.predict(input_df)
print("Predicted Marks:",prediction)
```

```python
# Prepare data for graph
assign_mean=df["Assignment"].mean()
study_range=np.linspace(df["Study Hours"].min(),df["Study Hours"].max(),100)
```

```python
# Predict marks for the graph
assign_mean=df["Assignment"].mean()
study_range=np.linspace(df["Study Hours"].min(),df["Study Hours"].max(),100)
```

```python
# Predict marks for the graph
marks_pred=model.predict(np.column_stack((study_range,np.full_like(study_range,assign_mean))))
```

```python
# Plot the graph
plt.scatter(df["Study Hours"],df["Marks"],color="blue",label="data")
plt.plot(study_range,marks_pred,color="red",label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
