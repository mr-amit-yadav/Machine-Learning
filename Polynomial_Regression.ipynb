## Polynomial Regression

Polynomial Regression is a type of regression that models the relationship between the independent variable X and dependent variable Y as nth-degree polynomial.

Formula: <br>
Y = b0  +  b1X  +  b2X2 + b3X3  + … + bnXn<br>
- When to use Polynomial regression<br>
- When data shows a curve pattern<br>
- When linear regression underfits <br>
- For trend forecasting, economics, weather pattern<br>


### A] Mean Squared Error (MSE)

Measures the average of squared differences between actual and predicted values.

Penalizes larger errors more due to squaring.

Formula:
$$
\text{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Lower MSE ⇒ Better model


### B] Mean Absolute Error (MAE)

Measures the average of absolute differences between actual and predicted values.

Formula:

$$
\text{MAE} = \frac{1}{n}\sum_{i=1}^{n}\lvert y_i - \hat{y}_i \rvert
$$


### C] Root Mean Squared Error (RMSE)

Square root of Mean Squared Error.
Gives error in the same units as the target variable.
Formula:
$$
\text{RMSE} = \sqrt{\text{MSE}}
$$

Helps understand error magnitude

### D] R² Score (Coefficient of Determination)

Measures how well the model explains the variance in the target variable.

Formula:

$$
R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}
$$


Interpretation:

R² = 1 → Perfect model

R² = 0 → Model predicts mean only

R² < 0 → Model worse than baseline

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Creating Dataset
X =np.linspace(0,10,50).reshape(-1,1)
Y=5+2*X + X**2 + np.random.randn(50,1)*5 #quatratic + noise
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Create the model
model = LinearRegression()
model.fit(X_poly,Y)
```

```python
# Predict Values
y_pred = model.predict(X_poly)
```

```python
# Evaluate model Performance
y_pred = model.predict(X_poly)
```

```python
# Evaluate Model Performance
print(f"MSE: {mean_squared_error(Y, y_pred)}")
print(f"R2 Score: {r2_score(Y, y_pred)}")
```

```python
# Visualization
plt.scatter(X,Y,label="Actual Data")
plt.plot(X,y_pred, color='red', linewidth=2, label="Polynomial Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()
```
