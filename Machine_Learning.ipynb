```python
# Importing the files
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Define Your Data
X =np.array([[1],[2],[3],[4],[5]]) # Study Hours
Y = np.array([10,20,30,40,50]) # Marks
```

```python
# Creating the model
model_simple = LinearRegression()
# Training the model
model_simple.fit(X,Y)
```

```python
# Visualization
plt.scatter(X,Y)
plt.plot(X,model_simple.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Simple Linear Regression: Study Hours Vs Marks")
plt.show()
```
